#Updating requested parameters in windows data

##Initial windows data

```
window_DF = pd.read_csv("windows.csv",sep=";",index_col=0,header=0) 
window_DF
```

##IAC calculation

###**Updating IAC_cl value**

For this, we want to read this IAC_cl values of each window, based on the Window_ID and IntShading_ID from the "IAC_cl.csv" file:

```
def IAC_cl_finder(row):
  window_ID = row["Window_ID"]
  intShading_ID = row["IntShading_ID"]
  IAC_cl_DF = pd.read_csv("IAC_cl.csv",sep=";",index_col=1,header=0)
  IAC_cl_value = IAC_cl_DF.loc[window_ID,intShading_ID]
  return IAC_cl_value
```

```
window_DF.loc[:,"IAC_cl"] = window_DF.apply(IAC_cl_finder,axis=1)
window_DF.loc[:,["Window_ID","IntShading_ID","IAC_cl"]]
```

###**IAC**

Using the formula: IAC = 1 + IntShading_closness*(IAC_cl-1)

```
window_DF.loc[:,"IAC"] = 1 + window_DF.loc[:,"IntShading_closeness"]*(window_DF.loc[:,"IAC_cl"]-1)
window_DF.loc[:,["IntShading_closeness","IAC_cl","IAC"]]
```

##PXI calculation

###**Ed (diffuse radiation)**

For this, we build a function to find the relevant value from the file "DiffuseIrradiance.csv". We considered the **lattitude of 45 degree** as for the city of Piacenza and reading the direction of each window from window data:

```
def Ed_finder(row):
  direction=row["Direction"]
  Ed_DF = pd.read_csv("DiffuseIrradiance.csv",sep=";",index_col=0,header=0)
  latitude = "45"
  Ed_value = Ed_DF.loc[direction,latitude]
  return Ed_value
```

```
window_DF.loc[:,"Ed"] = window_DF.apply(Ed_finder,axis=1)
window_DF.loc[:,["Direction","Ed"]]
```

###**ED (beam or direct radiation)**
Same procedure as of diffuse radiation finder:


```
def EB_finder(row):
  direction=row["Direction"]
  EB_DF = pd.read_csv("BeamIrradiance.csv",sep=";",index_col=0,header=0)
  latitude = "45"
  EB_value = EB_DF.loc[direction,latitude]
  return EB_value
```

```
window_DF.loc[:,"ED"] = window_DF.apply(EB_finder,axis=1)
window_DF.loc[:,["Direction","ED"]]
```

###**SLF**
Same procedure as of diffuse radiation finder:

```
def SLF_finder(row):
  direction = row["Direction"]
  lattitude = "45"
  SLF_DF = pd.read_csv("SLF.csv",sep=";",index_col=0,header=0)
  SLF_value = SLF_DF.loc[direction,lattitude]
  return SLF_value
```

```
window_DF.loc[:,"SLF"] = window_DF.apply(SLF_finder,axis=1)
window_DF.loc[:,["Direction","SLF"]]
```

###**Fshd**
Implementing the equation related to calculation of Fshd:

Fshd = min(1,max(0,(SLF*Doh-Xoh)/height))

```
for i in window_DF.index:
  window_DF.loc[i,"Fshd"] = min(1,max(0,(window_DF.loc[i,"SLF"]*window_DF.loc[i,"Doh"]-window_DF.loc[i,"Xoh"])/window_DF.loc[i,"Height"]))

window_DF.loc[:,["SLF","Doh","Xoh","Height","Fshd"]]
```

###**PXI**

Implementing the formula below, related to the calculation of PXI:

PXI = Tx(Ed + (1-Fshd)ED)

```
window_DF.loc[:,"PXI"] = window_DF.loc[:,"Tx"] * (window_DF.loc[:,"Ed"] + (1-window_DF.loc[:,"Fshd"])*window_DF.loc[:,"ED"])
window_DF.loc[:,["Tx","Ed","Fshd","ED","PXI"]]
```

##FFs
For updating the values of FFs, we should find the data from the file "FFs.csv" and extract relevant data based on the direction of each window and considering the the building is for a single family:


```
def FFs_finder(row):
  direction = row["Direction"]
  family = "SingleFamilyDetached"
  FFs_DF = pd.read_csv("FFs.csv",sep=";",index_col=0,header=0)
  FFs_value = FFs_DF.loc[direction,family]
  return FFs_value
```

```
window_DF.loc[:,"FFs"] = window_DF.apply(FFs_finder,axis=1)
window_DF.loc[:,["FFs"]]
```
