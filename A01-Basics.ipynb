#**Exercise A**

##Inputs


```
T_inf1 = 20 #[degreeC]
T_inf2 = -10 #[degreeC]
h1 = 10 #[W/m^2/degreeC]
h2 = 40 #[W/m^2/degreeC]
k_glass = 0.78 #[W/m/degreeC]
A = 0.8*1.5 #[m^2]
L = 0.008 #[m]
```

##Model

```
R_conv1 = 1/h1/A
R_glass = L/(k_glass*A)
R_conv2 = 1/h2/A
R_tot = R_conv1 + R_glass + R_conv2

Q_dot_tot = (T_inf1-T_inf2)/R_tot

T1 = T_inf1 - Q_dot_tot*R_conv1

print("Heat transfer rate is: " + str(Q_dot_tot) + " [W]")
print("Temperature of glass's inner surface is: " + str(T1) + " [degreeC]")
```

#**Exercise B**

##Thermal resistance of the glass

```
material = "glass"

if material == "glass":
  k = 0.78 #[W/m/degreeC]
elif material == "air":
  k = 0.026 #[W/m/degreeC]

L = 0.004 #[m]
A = 0.8*1.5 #[m^2]
R_glass = L/A/k

print(R_glass)
```

##Thermal resistance of air gap


```
material = "air"
if material == "glass":
  k = 0.78 #[W/m/degreeC]
elif material == "air":
  k = 0.026 #[W/m/degreeC]

L = 0.01 #[m]
A = 0.8*1.5 #[m^2]

R_AG = L/A/k

print(R_AG)
```

##Thermal resistance of the inner surface with free stream

```
h_conv1 = 10 #[W/m^2/degreeC]
A = 0.8*1.5 #[m^2]

R_conv1 = 1/h_conv1/A

print(R_conv1)
```

##Thermal resistance of the outer surface with free stream

```
h_conv2 = 40 #[W/m^2/degreeC]
A = 0.8*1.5 #[m^2]

R_conv2 = 1/h_conv2/A

print(R_conv2)
```

##Total thermal resistance

```
R_total = R_conv1 + R_glass + R_AG + R_glass + R_conv2

print("Total thermal resistance of the window is: " + str(R_total) + " [degreeC/W]")
```

##Heat transfer rate

```
T_inf1 = 20 #[degreeC]
T_inf2 = -10 #[degreeC]

Q_dot_total = (T_inf1 - T_inf2) / R_total

print("The heat transfer rate is: " + str(Q_dot_total) + " [W]")
```

##Inner glass's surface temperature

```
T1 = T_inf1 - Q_dot_total*R_conv1

print("The inner glass's surface temperature is: " + str(T1) + " [degreeC]")
```
