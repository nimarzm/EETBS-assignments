#Example B

##Inputs

```
T_inf1 = 20 #[degreeC]
T_inf2 = -10 #[degreeC]
A = round(0.8*1.5,2) #[m^2]
```

###Thermal resistance calculation

```
R1 = ['R_in','conv',10,A]
R2 = ['R_g','cond',0.004,0.78,A]
R3 = ['R_AG','cond',0.01,0.026,A]
R4 = R2.copy()
R5 = ['R_out','conv',40,A]
R_list = [R1,R2,R3,R4,R5]
R_list
```

```
R_tot = 0
for R in R_list:
  if R[1] == 'cond':
    Rvalue = R[2]/R[3]/R[4]
    R.append(Rvalue)
  elif R[1] == 'conv':
    Rvalue = 1/R[2]/R[3]
    R.append(Rvalue)
  R_tot += Rvalue
```

##Outputs

```
print('Total thermal resistance is equal to ' + str(round(R_tot,4)) + ' [degreeC/W].')
```

```
Q_dot = (T_inf1-T_inf2)/R_tot
print('Steady rate of the heat transfer through this wall is equal to ' + str(round(Q_dot,2)) + ' [W].')
```

```
T1 = T_inf1 - Q_dot*R1[-1]
print("The temperature of the window's inner surface is equal to " + str(round(T1,1)) + ' [degreeC].')
```
