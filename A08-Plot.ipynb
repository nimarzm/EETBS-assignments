```
import matplotlib.pyplot as plt
```

##Defining functions

```
def pieHeat(heating_load):
  """This function makes pie chart, by taking the heating load in our-class-defined order!"""
  heating = plt.figure()
  heating_share = ["Opaque","Fenestration","Infiltration & Ventilation","Distribution Loss"]
  heating_color = [(1,0.2,0),(0.8,0.2,0),(0.6,0.2,0),(0.4,0.2,0)]
  plt.pie(heating_load, labels=heating_share, colors=heating_color, startangle=90, autopct="%1.1f%%", radius=2.5, counterclock=False)
  return
```
