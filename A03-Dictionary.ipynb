##**Part 1:** Double pane window

```
T_inf1 = 20 #[degreeC]
T_inf2 = -10 #[degreeC]
A = round(0.8*1.5,2)
```

```
R1 = {'name':'R_in','type':'conv','h':10,'area':A}
R2 = {'name':'R_glass','type':'cond','length':0.004,'k':0.78,'area':A}
R3 = {'name':'R_AG','type':'cond','length':0.01,'k':0.026,'area':A}
R4 = R2
R5 = {'name':'R_out','type':'conv','h':40,'area':A}

ResistanceList = [R1,R2,R3,R4,R5]
```

```
Rtot = 0
for anyR in ResistanceList:
  if anyR['type'] == 'conv':
    anyR['Rvalue'] = round(1/anyR['h']/anyR['area'],4)
  elif anyR['type'] == 'cond':
    anyR['Rvalue'] = round(anyR['length']/anyR['k']/anyR['area'],4)
  Rtot += anyR['Rvalue']
```

```
print("The value of total thermal resistance is: " + str(round(Rtot,4)) + " [degreeC/W]")
```

```
Q_dot_tot = round((T_inf1-T_inf2)/Rtot,2)
print("The steady rate of heat transfer is: " + str(Q_dot_tot) + " [W]")
```

```
T1 = T_inf1 - Q_dot_tot*R1['Rvalue']
print("The temperature of window's inner surface is: " + str(round(T1,1)) + " [degreeC]")
```

##**Part 2:** Simplified wall calculation

```
ThermalResistanceLib = {'OutsideAir':{"R'":0.03},'WoodFiberboard':{"R'":0.23,'StandardLength':13},'WoodBevelLS':{"R'":0.14,'StandardLength':13},
                        'GlassFiber':{"R'":0.7,'StandardLength':25},'WoodStud':{"R'":0.63,'StandardLength':90},'GypsumBoard':{"R'":0.079,'StandardLength':13},
                        'InsideAir':{"R'":0.12}}
```

```
Rp1 = {'Name':'OutsideAir','type':'conv'}
Rp2 = {'Name':'WoodBevelLS','type':'cond','length':13}
Rp3 = {'Name':'WoodFiberboard','type':'cond','length':13}
Rp4 = {'Name':'GlassFiber','type':'cond','length':90}
Rp5 = {'Name':'WoodStud','type':'cond','length':90}
Rp6 = {'Name':'GypsumBoard','type':'cond','length':13}
Rp7 = {'Name':'InsideAir','type':'conv'}
ResistanceList2 = [Rp1,Rp2,Rp3,Rp4,Rp5,Rp6,Rp7]
```

```
i=0
for anyRp in ResistanceList2:
  i+=1
  if anyRp['type'] == 'cond':
    for anyobj in ThermalResistanceLib:
      if anyobj == anyRp['Name']:
        anyRp["R'"] = round(ThermalResistanceLib[anyobj]["R'"]*anyRp['length']/ThermalResistanceLib[anyobj]['StandardLength'],3)
  elif anyRp['type'] == 'conv':
    for anyobj in ThermalResistanceLib:
      if anyobj == anyRp['Name']:
        anyRp["R'"] = round(ThermalResistanceLib[anyobj]["R'"],3)
```

```
ResistanceList2
```

```
Rp_tot1 = 0
for anyRp in ResistanceList2:
  if anyRp['Name'] != 'WoodStud':
    Rp_tot1 += anyRp["R'"]
print("The total unit thermal resistance of INSULATION PART is: " + str(Rp_tot1) + " [m^2 degreeC/W]")
```

```
Rp_tot2 = 0
for anyRp in ResistanceList2:
  if anyRp['Name'] != 'GlassFiber':
    Rp_tot2 += anyRp["R'"]
print("The total unit thermal resistance of WOOD-STUD part is: " + str(Rp_tot2) + " [m^2 degreeC/W]")
```

```
U_tot = 0.75/Rp_tot1 + 0.25/Rp_tot2
Rp_tot = 1/U_tot
print("The overall unit thermal resistance (the R-value) is: " + str(round(Rp_tot,4)) + " [m^2 degreeC/W]")
print("The overall heat transfer coefficient (the U-factor) is: " + str(round(U_tot,4)) + " [W/m^2/degreeC]")
```

```
A_tot = 0.8*50*2.5
Q_dot_tot = U_tot * A_tot * (22-(-2))
print("The rate of heat loss through the wall is: " + str(round(Q_dot_tot,1)) + " [W]")
```
